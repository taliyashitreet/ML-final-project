# An explanation of each feature
*  App	- the name of the Application
* Category- Category the app belongs to
* Rating - Overall user rating of the app 
* Reviews - Number of user reviews for the app
* Size - Size of the app	
* Installs - Number of user downloads	
* Type - Paid or Free	
* Price - How much does the app cost	
* Content Rating- What is the target audience
* Genres - Category the app belongs to
* Last Updated - When was the latest version update
* Current Ver	- current version of this app
* Android Ver - version of this app in Android

```python
gps.isnull().sum()
```

```python
#cleaning of rating data:
NotNull=gps[pd.notnull(gps['Rating'])]

#lets check mean and median
mean=np.mean(NotNull['Rating'])
median=np.median(NotNull['Rating'])

print(mean,median)
```

```python
# I prefer to fill the Null's Rating with the median
gps['Rating'].fillna(median,inplace=True)
# In other features missing value % is not considrable so drop Null
gps.dropna(inplace=True)
```

```python
# Remove dublicate values
gps.duplicated().value_counts()
```

```python
#We have seen that there are 483 duplicates of rows in the data frame, so we can drop it
gps.drop_duplicates(inplace=True) 
```

```python
#the column Genres, doesn't give us anything new that we can't understand from the Category column
gps.drop('Genres',axis='columns', inplace=True)
```

```python
# Looking for apps, where 'Price' is represented with non-decimal value
gps[gps['Price'].str.replace('$','').str.replace('.','').str.isnumeric()]
```

```python
# Checking values in Installs column
gps['Installs'].unique()
```

#### We can remove the Plus and comma signs 

```python
# Checking values in Installs column
gps['Size'].unique()
```

### In Sizes we have:
* values marked with 'M', that I'll convert into numerical millions
* values marked with 'k', that I'll replace with thousands
* values marked with 'Varies with device', there are apps that vary in size from device to device

```python
# Checking values in price column
gps['Price'].unique()
```

* I was trynig to convert column Price to numeric, but I was getting an error becouse of  'Everyone'.
* It seems that record with index 10472 have errors in multimple columns so the best idea is to drop this row

```python
gps=gps.drop(gps.index[10472])
```

```python
# Now we can cange it to numeric
gps['Reviews']=pd.to_numeric(gps.Reviews)
gps = gps.sort_values(by='Reviews', ascending=False)
```

```python
# Changing types of columns as described above.
gps['Price'] = gps['Price'].str.replace('$', '').astype('float')
gps['Installs'] = gps['Installs'].str.replace(',', '').str.replace('+', '').astype('int')
gps['Size'] = gps['Size'].str.replace('M', 'e+6').str.replace('k', 'e+3').str.replace('Varies with device', '0').astype('float')
```

```python
gps.head(15)
```

# Vizualization

```python
plt.figure(figsize=(14,8))
plt.title("Total Apps per Category in Play Store")
sns.countplot(y='Category',data=gps,order=gps['Category'].value_counts().iloc[:20].index);
```

* we can inffer that for "family", "gams" and "tools" there are the most apps

```python
piv_cat = gps.groupby('Category')
x = piv_cat['Installs'].agg(np.sum)
plt.figure(figsize=(15,5))
plt.plot(x,color='red')
plt.xticks(rotation=90)
plt.ylabel('Installs')
plt.title('Total installs per category')
plt.show()
```

```python
plt.figure(figsize=(10,10))
explode=[0.1,0]
gps['Type'].value_counts().plot(kind='pie',autopct="%2i%%",explode=explode)
plt.legend()
plt.show()
```
